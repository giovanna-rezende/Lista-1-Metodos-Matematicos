```
import numpy as np
from math import sqrt
import math as m
import sympy
from matplotlib import pyplot
```

```
# Seção 7-5, exercício 7a
# x = (1,2,0), y = (-2,0,-2), z = (1,1,6)

x = np.array([1,2,0])
y = np.array([-2,0,-2])
z = np.array([1,1,6])
# Cálculo do vetor normal ao plano
n = np.cross(y,z)
# Cálculo da distância do vetor x ao plano, usando a fórmula descrita no caderno
d = np.dot(n,x-y)/sqrt(np.dot(n,n))
print("A distância entre o vetor x e o plano é d =", d)
```

```
# Seção 9-4, exercício 5
# Mostrando série de Fourier calculada

eixo_x = []
eixo_y = []
x = -m.pi
while x <= m.pi:
  y = 1/4*m.pi - 1/2
  for k in range(1,5): # Calculando termos da série
    y += (sin(k/2)/k*m.pi)*cos(k*x) + (pow(-1,k+1)*2/k + cos(k/2)/k*m.pi - 2*sin(k/2)/pow(k,2)*m.pi)*sin(k*x)
  eixo_x.append(x)
  eixo_y.append(y)
  x += 0.2

pyplot.plot(eixo_x, eixo_y)
pyplot.xlabel("x")
pyplot.ylabel("y")
pyplot.title("Série de Fourier")
pyplot.grid("True")
```
